# Gráficos no Python utilizando o pacote matplotlib

```
import matplotlib.pyplot as plt
```

```
plt.plot([1, 2, 3, 4], 'ro')
plt.show()
```

```
import numpy as np
x = np.arange(0, 10, 0.1)

# criar o subgráficos
fig1, f1_axes = plt.subplots(ncols=2, nrows=2, figsize=(15, 10))

# título na figura 1
fig1.suptitle('Minha primeira figura', size=30)

# gráfico 1
f1_axes[0, 0].plot(np.sin(x), label='sen(x)', color='black')
# título
f1_axes[0, 0].set_title('Caixa 1')
f1_axes[0, 0].plot(np.cos(x), label='cos(x)', color='red')
# nomeando os eixos
f1_axes[0, 0].set_xlabel('Caixa 1 - eixo x')
f1_axes[0, 0].set_ylabel('Caixa 1 - eixo y')
#legenda - precisa adicionar a label nos plots
f1_axes[0, 0].legend()

f1_axes[0, 1].plot(np.cos(x))
f1_axes[0, 1].set_title('Caixa 2')

f1_axes[1, 0].plot(np.tan(x))
f1_axes[1, 0].set_title('Caixa 3')

f1_axes[1, 1].set_title('Caixa 4')
```
